```python
import pandas as pd
```

```python
datos=pd.read_csv("melb_data.csv")
datos
```

```python
datos.describe()
```

```python
datos=datos.dropna(axis=0)
datos
```

Machine learning

```python
y= datos['Price']
x= datos[['Rooms','Bathroom','Landsize','Lattitude','Longtitude']]
```

```python
from sklearn.tree import DecisionTreeRegressor
modelo=DecisionTreeRegressor()
modelo.fit(x,y)
```

```python
print('Prediciones de las 5 casas')
print(x.head())
print('El precio real es :')
print(y.head())
print('Las predicciones son:')
print(modelo.predict(x.head()))
```

MIN ABSOLUT ERROR

```python
from sklearn.metrics import mean_absolute_error
prediccion_prices=modelo.predict(x)
mean_absolute_error(y,prediccion_prices)
```

Validation

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y =train_test_split(x,y)
modelo= DecisionTreeRegressor()
modelo.fit(train_x,train_y)
val_prediccion= modelo.predict(val_x)
print(mean_absolute_error(val_y,val_prediccion))
```

```python
def get_mae(max_leaf_nodes,train_x,val_x,train_y,val_y):
    model=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes)
    model.fit(train_x,train_y)
    predict_val=model.predict(val_x)
    mae=mean_absolute_error(val_y,predict_val)
    return(mae)
```

```python
for max_leaf_nodes in [5,50,500,5000]:
    my_mae=get_mae(max_leaf_nodes, train_x,val_x,train_y,val_y)
    print("Max leaf nodes:%d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_x,train_y)
melb=forest_model.predict(val_x)
print(mean_absolute_error(val_y,melb))
```
