student's performace

```python
import pandas as pd
```

```python
students_p= pd.read_csv('study_performance.csv')
students_p
```

pre-processing
Prediction variable :score

```python
score_columns= [columna for columna in students_p.columns if columna.endswith('score')]
score_columns
```

```python
students_p['score']= round(students_p[score_columns].sum(axis=1)/30)
students_p
```

```python
students_p.drop(columns=score_columns, inplace= True)
students_p
```

Variables categoricas

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder= OneHotEncoder(sparse_output=False)
```

```python
encoded_data=encoder.fit_transform(students_p[['gender']])
encoded_data
```

```python
encoded_df=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['gender']))
encoded_df
```

```python
students_p=pd.concat([students_p,encoded_df],axis=1)
students_p.drop('gender', axis=1, inplace=True)
students_p
```

```python
encoder= OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(students_p[['race_ethnicity']])
encoded_df=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['race_ethnicity']))
students_p=pd.concat([students_p,encoded_df],axis=1)
students_p.drop('race_ethnicity', axis=1, inplace=True)
students_p
```

```python
encoder= OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(students_p[['lunch']])
encoded_df=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['lunch']))
students_p=pd.concat([students_p,encoded_df],axis=1)
students_p.drop('lunch', axis=1, inplace=True)
students_p
```

```python
encoder= OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(students_p[['test_preparation_course']])
encoded_df=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['test_preparation_course']))
students_p=pd.concat([students_p,encoded_df],axis=1)
students_p.drop('test_preparation_course', axis=1, inplace=True)
students_p
```

Ordinal encoding

```python
students_p['parental_level_of_education'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_categories= ["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"]
encoder= OrdinalEncoder(categories=[ordinal_categories])
encoded_data=encoder.fit_transform(students_p[['parental_level_of_education']])
df_encoded=pd.DataFrame(encoded_data, columns=['parental_level_of_education_encoder'])
students_p=pd.concat([students_p,df_encoded],axis=1)
students_p.drop('parental_level_of_education', axis=1, inplace=True)
students_p
```

Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
x= students_p.drop('score',axis=1)
y=students_p['score']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=21)
x_train
```

```python
rf_model=RandomForestRegressor(n_estimators= 100, random_state=42)
rf_model.fit(x_train, y_train)

lr_model= LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
rf_prediction=rf_model.predict(x_test)
lr_prediction=lr_model.predict(x_test)
```

```python
rf_mse= mean_absolute_error(y_test, rf_prediction)
lr_mse=mean_absolute_error(y_test, lr_prediction)
print("Random Forest MSE : ", rf_mse)
print("Linear Regression MSE: ", lr_mse)
```
